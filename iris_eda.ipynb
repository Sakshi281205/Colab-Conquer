```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("https://bit.ly/4nejNue")

# Convert specific columns to float
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for col in col_names:
    data[col] = data[col].astype(float)

# Print first few rows
print("üîç Dataset Preview:\n", data.head())

# Basic Statistics
print("\nüìà MEAN:\n", data[col_names].mean())
print("\nüìä MEDIAN:\n", data[col_names].median())
print("\nüìå MODE:\n", data[col_names].mode().iloc[0])  # Show first mode row
print("\nüìâ STANDARD DEVIATION:\n", data[col_names].std())
print("\nüìè VARIANCE:\n", data[col_names].var())
print("\n‚¨ÜÔ∏è MAX:\n", data[col_names].max())
print("\n‚¨áÔ∏è MIN:\n", data[col_names].min())
print("\nüßæ DESCRIBE:\n", data.describe())

# Visualize Mean, Median, Std on a Line Plot
plt.figure(figsize=(10, 6))
plt.plot(col_names, data[col_names].mean(), label='Mean', marker='o')
plt.plot(col_names, data[col_names].median(), label='Median', marker='s')
plt.plot(col_names, data[col_names].std(), label='Standard Deviation', marker='^')
plt.title("üìä Iris Feature Statistics")
plt.xlabel("Features")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Histograms
plt.figure(figsize=(12, 8))
for i, col in enumerate(col_names):
    plt.subplot(2, 2, i + 1)
    plt.hist(data[col], bins=15, color='skyblue', edgecolor='black')
    plt.title(f'{col.capitalize()} Histogram')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.suptitle("üîç Feature Distributions", fontsize=16, y=1.02)
plt.show()
```
